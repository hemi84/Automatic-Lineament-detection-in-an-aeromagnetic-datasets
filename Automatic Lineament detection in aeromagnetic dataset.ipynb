```python
import cv2
import numpy as np
import pandas as pd
```

```python
img = cv2.imread('tilt2.png')
org = cv2.imread('org1.png')
```

```python
import hemi_fault_detection2 as hf
```

```python
lines = hf.detect_fault(img)
```

```python
for points in lines:
    x1,y1,x2,y2 = points[0]
    cv2.line(org, (x1,y1),(x2,y2),(0,0,0))
    cv2.line(img, (x1,y1),(x2,y2),(0,0,0))
```

```python
cv2.imwrite('faultst.png', org)
cv2.imwrite('faultso.png', img)
```
